```python
import pandas as pd
import numpy as np
from keras.layers import Dense, Activation, Dropout
from keras.models import Sequential
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
X = pd.read_csv("input 1(198143).csv")
X
```

```python
y = pd.read_csv("output 1(198143).csv")
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 0)
X_train
```

```python
Rec = 4200
Ree = 180
v = 1/Rec
Ut = Ree/Rec
U2 = v/(Ut*Ut)

X_train = X_train * U2
X_test  = X_test * U2
X       = X * U2


y_train = y_train/(Ut*Ut)
y_test  = y_test/(Ut*Ut)
y       = y/(Ut*Ut)
```

```python
model = Sequential()
model.add(Dense(128, activation = 'tanh', input_dim = 9))
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dense(units = 1, activation = 'linear'))

model.compile(optimizer = 'adam', loss = 'Huber')

history = model.fit(X_train, y_train, validation_split = 0.1, batch_size = 256, epochs = 500)
```

```python
#input to hidden
layer_1_weights = model.layers[0].get_weights()[0]
layer_1_biases = model.layers[0].get_weights()[1]


#hidden to hidden
layer_2_weights = model.layers[1].get_weights()[0]
layer_2_biases = model.layers[1].get_weights()[1]

#hidden to hidden
layer_3_weights = model.layers[2].get_weights()[0]
layer_3_biases = model.layers[2].get_weights()[1]

#hidden to output
layer_4_weights = model.layers[3].get_weights()[0]
layer_4_biases = model.layers[3].get_weights()[1]
```

```python
def relu(x):
   return np.maximum(0,x)


def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x

def tanh(x):
  t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  return t
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler



inputs = np.transpose(X_test)
Wh1 = np.transpose(layer_1_weights)
bh1 = layer_1_biases.reshape(128,1)
Z1 = np.matmul(Wh1, inputs) + bh1
h1 = tanh(Z1)

Wh2 = np.transpose(layer_2_weights)
bh2 = layer_2_biases.reshape(128,1)
Z2 = np.matmul(Wh2, h1) + bh2
h2 = relu(Z2)

Wh3 = np.transpose(layer_3_weights)
bh3 = layer_3_biases.reshape(128,1)
Z3 = np.matmul(Wh3, h2) + bh3
h3 = relu(Z3)

Wh4 = np.transpose(layer_4_weights)
bh4 = layer_4_biases.reshape(1,1)
Z4 = np.matmul(Wh4, h3) + bh4
h4 = linear(Z4)

output = np.transpose(h4)


y_test_arr = y_test.to_numpy()
y_test_arr = y_test_arr.reshape((29722, 1))
mse_test = mean_squared_error(y_test_arr,output)
print(mse_test)
```

```python
y_pred = model.predict(X_test)
mse_test = mean_squared_error(y_test_arr,y_pred)
print(mse_test)
```
